# Run/plot different feature elimination (FE) methods
* There are pros/cons for different FE methods, and I am agnostic right for now
* Using the union of selected features from several approaches, can yield a 'short list' of potentially important variables

<B> SHAP based FE </B> <br>
* Rank features in the order of importance in the initial dataset and consequtively elimiate features (select best performing feature subset based on $R^{2}$ with 5-fold cross-validation)
    
<B> SHAP based RFECV </B><br>
* Similar to SHAP based FE, but SHAP-based feature importance is calculated at each step of the elimination process.
    
<B> Boruta based FE </B><br>
* Similar to SHAP based FE, but initial feature importance ranking is determined by the Boruta method; creating shadow features (shuffled feature sets) and determining which features are more important than the shadow features. Repeating this x times yields a ranking of features by how often they were more important than chance. After obtaining this ranking, features were consequtively elimiated. The best performing feature subset based on $R^{2}$ with 5-fold cross-validation was selected.

```python
try: 
    # Assess if feature elimination methods were already run: 
    # Note:  outputs of FE methods are stored in the *_outputs properties as a dictionaries so they can be viewed later
    assert type(FE.SHAP_based_FE_outputs)==dict  
    assert type(FE.SHAP_based_RFECV_outputs)==dict
    assert type(FE.Boruta_based_FE_outputs)==dict
except: 
    start = time.time()
    n_steps=16 # Number of feature elimination steps to run (max would be p-1, where p==number of features)
    verbose = False
    print("n_steps: ", n_steps) # I used exponental spacing so more features are eliminated early
    FE.run_SHAP_based_FE(n_steps=n_steps, verbose=verbose);
    FE.run_SHAP_based_RFECV(n_steps=n_steps, verbose=verbose, max_evals=0); #max_evals: hyperparameter tuning at each iteration
    FE.run_boruta_FE(n_steps=n_steps, verbose=verbose, max_iter=100) # max_iter: Number models to use for feature rankings
    FE.get_union_features()
    FE.save()
    end = time.time()
    print("Execution time: {:.2f}s".format(np.round(end - start,5)))
FE.plot_FE_comparison()
# Features selected by different models:
for d in [FE.Boruta_based_FE_outputs, FE.SHAP_based_FE_outputs, FE.SHAP_based_RFECV_outputs]:
    print(d['FE_method'],':', d['optimal_features'],'\n')
try:
    assert type(FE.union_features) == list
    print('Union of selected features orederd by importance:')
    print(FE.union_features)
except:
    print('Ordering union of selected features by importance:')
    print(FE.get_union_features(importance_sort=True))
```

# Analysis 
* Conducted using the union of automatically selected features to ensure important covariates are included
* (Domain knowledge and correlation analysis may also help decide what features to include/exclude)

```python
mdlVars = copy.deepcopy(FE.union_features)
mdlVars.append(target)
print(mdlVars)
try:
    fileName = max(glob.glob(outputs_dir + target+ "_analysis*.p"), key=os.path.getctime)
    print("Loaded:", fileName)
    analysis= pickle.load(open(fileName, 'rb'))
    analysis.shap_summary_plots()
except:
    analysis = xgb_shap(df= df[mdlVars], target = target, bootstrap_iterations=500, max_evals=25, outputs_dir=outputs_dir)
    analysis.tune_model() # Tune hyperparameters for the model 
    analysis.shap_summary_plots() # Calling this will automatically perform SHAP analysis
    # Generate bootstrapped estimates of SHAP values to provide confidence intervals around estimates:
    analysis.run_bootstrap_SHAP_analysis_parallel(); # Note: this can be run on the cluster
    # Generate interaction values (to look for potential interactions between covariates):
    analysis.shap_interaction_values;
    analysis.save(name='analysis') # Save analysis to outputs directory. 
```
